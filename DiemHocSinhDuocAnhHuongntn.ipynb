```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('stud_perf_exam.csv')
df
```

```python
#Columns ion dataset
print(df.columns)
print(f"Dataset có {len(df.columns) } cột")
```

```python
df.rename(columns={'gender':'giới tính'},inplace=True)
df.rename(columns={'race/ethnicity':'chủng tộc hoặc dân tộc sinh viên'},inplace=True)
df.rename(columns={'parental level of education':'trình độ học vấn của phụ huynh'},inplace=True)
df.rename(columns={'lunch':'bữa trua(có được hỗ trợ hay không)'},inplace=True)
df.rename(columns={'test preparation course':'tham gia khóa học ôn thi'},inplace=True)
df.rename(columns={'math score':'Điểm toán'},inplace=True)
df.rename(columns={'reading score':'Điểm đọc'},inplace=True)
df.rename(columns={'writing score':'Điểm ghi'},inplace=True)
df
```

```python
df.info()
```

```python
print(f"có {df.isna().sum().sum()} giá trị là NaN")
```

```python
cat_col=[]
for col in df.columns:
    if(df[col].dtypes=='object'):
        cat_col.append(col)
print(f"Có tổng cộng {len(cat_col)} cột dữ liệu phân loại trong dataset")
print(cat_col)
```

```python
numberical=[]
for col in df.columns:
    if df[col].dtypes!='object':
        numberical.append(col)
print(f"Có tổng cộng {len(numberical)} cột có số trong dataset")
print(numberical)
```

```python
df[cat_col].dropna()#loại bỏ NaN trong cột dữ liệu phân loại
```

```python
plt.figure(figsize=(12,4),facecolor='#99A3A4')#màu nền và kích thước
for col,i in zip(cat_col,range(len(cat_col))):
    plt.subplot(2,3,i+1)#2 dòng 3 cột, i+1 là vị trí thứ 1(tạo ra các đồ thị nhỏ)
    value=df[col].value_counts()#những giá trị đó xuất hiện bao nhiêu lần
    key=df[col].value_counts().index#cột đó gồm những giá trị nào
    plt.pie(value,labels=key,autopct="%.0f%%")#làm tròn(chữ số sau dấu phẩy là 0)
    plt.title(col)
    print("")
```
